```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
data_file = pd.read_csv("test.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split
labels = data['label']
col = list(data.columns)
col.remove('label')
data_train,data_test,label_train,label_test = train_test_split(data[col],labels,train_size=0.9)
data_train.head()
```

```python
print(data['label'].value_counts())
type(data_train)
```

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(data_train,label_train)
pred = rf.predict(data_test)
pred
```

```python
acc = accuracy_score(label_test,pred)
fs = f1_score(label_test,pred,average=None)
print(acc)
fs
```

```python
param_grid = { "criterion" : ["gini", "entropy"], "min_samples_leaf" : [1, 5, 10, 25, 50], "min_samples_split" : [2, 4, 10, 12, 16, 18, 25], "n_estimators": [50,100, 300]}
from sklearn.model_selection import GridSearchCV
rf = RandomForestClassifier(n_estimators=100, max_features='auto', oob_score=True, random_state=1, n_jobs=-1)
clf = GridSearchCV(estimator=rf, param_grid=param_grid, n_jobs=-1)
clf.fit(data[col], data['label'])
clf.best_params_
```

```python
rf = RandomForestClassifier(n_estimators =100)
rf.fit(data[col],labels)
pred = rf.predict(data_file)
pd.DataFrame({"ImageId":list(range(1,len(data_file)+1)),"Label":pred}).to_csv("output_svm.csv",index=False)
```
